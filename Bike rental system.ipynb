```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


class Analysis:
    dataset1 = pd.read_csv("C:\\Users\\ssuun\\PycharmProjects\\pythonProject3\\bikes.csv")
    df1 = pd.DataFrame(dataset1)
    dataset2 = pd.read_csv("C:\\Users\\ssuun\\PycharmProjects\\pythonProject3\\users.csv")
    df2 = pd.DataFrame(dataset2)
    new_df = df2.merge(df1, on="bike_id")  # merge two dataset
    dataset3 = pd.read_csv("C:\\Users\\ssuun\\PycharmProjects\\pythonProject3\\rentedBike.csv")
    df3 = pd.DataFrame(dataset3)
    # print("\n======================================================")
    # print("  BIKE Dataset ")
    # print("\n======================================================")
    # print(df1)
    # print("\n======================================================")
    # print("  USER Dataset ")
    # print("\n======================================================")
    # print(df2)
    # print("\n======================================================")
    # print("  RENTED BIKE Dataset ")
    # print("\n======================================================")
    # print(df3)

    #print(df1.dtypes)
    #print(df1.isnull())
    #print(df3.isnull())


    # Visualization between bike_type and location
    def bar_chart1(self):
        x = self.df1.location
        y = self.df1.type
        plt.bar(x, y)
        plt.xlabel("location")
        plt.ylabel("Bike type")
        plt.show()

    # Visualization between Bike_type and User_age
    def bar_chart2(self):
        x1 = self.new_df.type
        y1 = self.new_df.age
        plt.bar(x1, y1)
        plt.title("User age vs Bike")
        plt.xlabel("Bike type")
        plt.ylabel("user age")
        plt.show()

    # Visualization for Bike_type and season

    def bar_chart3(self):
        fig,ax=plt.subplots(figsize=(15,8))
        sns.set_style('white')
        #Bar plot for seasonwise Bike type distribution of counts
        sns.barplot(x='season',y='type',data=df3[['type','season']],hue='season',ax=ax)
        ax.set_title('Seasonwise Bike type distribution of counts')
        plt.show()
    
    #Visualization  User_age distribution of season:
    def bar_chart4(self):
        fig,ax1=plt.subplots(figsize=(15,8))
        #Bar plot for user_age distribution of season
        sns.barplot(x='season',y='age',data=df3[['season','age']],ax=ax1)
        ax1.set_title('Season wise distribution of User age')
        plt.show()

    def bar_chart5(self):
        x1 = self.df3.type
        y1 = self.df3.rental_days
        colors = ["maroon"]
        plt.bar(x1, y1,color = colors)
        plt.title("Bike_type vs Rental_days")
        plt.xlabel("Bike type")
        plt.ylabel("Rental_days")
        plt.show()
        
        #Heat map for correlation matrix of attributes
    def bar_chart6(self):
        correMtr=df3[["rental_days","age"]].corr()
        mask=np.array(correMtr)
        mask[np.tril_indices_from(mask)]=False
        fig,ax=plt.subplots(figsize=(15,8))
        sns.heatmap(correMtr,mask=mask,vmax=0.8,square=True,annot=True,ax=ax)
        ax.set_title('Correlation matrix of attributes')
        plt.show()
        
    def bar_chart7(self):
        data = {'type': ["Mountain Bike", "City Bike", "Electric Bike", "Road Bike", "Cruiser Bike", "Mountain Bike", "City Bike", "Electric Bike", "Road Bike", "Cruiser Bike", "Mountain Bike", "City Bike", "Electric Bike", "Road Bike", "Electric Bike", "Road Bike", "Cruiser Bike", "Mountain Bike", "Mountain Bike", "City Bike", "Mountain Bike", "Mountain Bike", "Mountain Bike", "Cruiser Bike", "Mountain Bike", "Mountain Bike", "City Bike", "Mountain Bike", "Mountain Bike", "Mountain Bike", "Mountain Bike", "City Bike", "Mountain Bike", "City Bike", "Electric Bike", "Road Bike", "Cruiser Bike", "Mountain Bike"],
        'Available': [5, 10, 13, 15, 5, 3, 0, 19, 22, 1, 0, 20, 27, 12, 7, 4, 9, 6, 4, 2, 3, 1, 13, 5, 6, 1, 0]}
        df = pd.DataFrame(data)
        # Calculate the total availability for each bike type
        availability_by_type =self.df.groupby('type')['Available'].sum().reset_index()
        # Create the pie chart
        plt.figure(figsize=(8, 8))
        plt.pie(availability_by_type['Available'], labels=availability_by_type['type'], autopct='%1.1f%%', startangle=140)
        plt.title("Availability by Bike Type")
        plt.show()

        

class Bike:
    def __init__(self, bike_id, type, location, is_available=True):
        self.bike_id = bike_id
        self.type = type
        self.location = location
        self.is_available = is_available
        self.daily_rate = 0  # Default daily rate

    def set_daily_rate(self, rate):
        self.daily_rate = rate

    def get_daily_rate(self):
        return self.daily_rate


class User:
    def __init__(self, user_id, password, age):
        self.user_id = user_id
        self.password = password
        self.age = age


bikes = []
users = []


def add_user(user, password, age):
    with open('users.csv', 'a') as file:
        file.write('\n' + user + ',')
        file.write(password + ',')
        file.write(age)

    user = User(user, password, age)
    users.append(user)
    print(f"User ({user.user_id}) added to the rental system.")


def get_user(user_id):
    for user in users:
        if user.user_id == user_id:
            return user
    return None


def add_bike(bike_id, type, location, Available, daily_rate):
    with open('bikes.csv', 'a') as file:
        file.write('\n' + bike_id + ',')
        file.write(type + ',')
        file.write(location + ',')
        file.write(Available + ',')
        file.write(str(daily_rate))

    bike = Bike(bike_id, type, location, Available)
    bike.set_daily_rate(daily_rate)
    bikes.append(bike)
    print(f"Bike {bike_id} ({type}) added to the rental system with a daily rate of {daily_rate:.2f}/-Rs.")


def remove_bike(bike_id):
    for bike in bikes:
        if bike.bike_id == bike_id:
            bikes.remove(bike)
            print(f"Bike {bike_id} removed from the rental system.")
            return
    print(f"Bike {bike_id} not found in the rental system.")


def rent_bike(user_id, bike_id, rental_days, age, season, bikeType):

    with open('rentedBike.csv', 'a') as file:
        file.write('\n' + bike_id + ',')
        file.write(user_id + ',')
        file.write(str(rental_days) + ',')
        file.write(age + ',')
        file.write(season + ',')
        file.write(bikeType)

    for bike in bikes:
        if bike.bike_id == bike_id and bike.is_available:
            user = get_user(user_id)
            if user:
                total_cost = bike.get_daily_rate() * rental_days
                bike.is_available = False
                print(f"{user_id} rented Bike {bike_id} for {rental_days} days. Total cost: {total_cost:.2f}/-Rs")
                return
            else:
                print(f"User {user_id} not found.")
                return
    print(f"Bike {bike_id} is not available for rent.")


def return_bike(user_id, bike_id):
    for bike in bikes:
        if bike.bike_id == bike_id and not bike.is_available:
            user = get_user(user_id)
            if user:
                bike.is_available = True
                print(f"{user.name} returned Bike {bike_id}.")
                return
            else:
                print(f"User {user_id} not found.")
                return
    print(f"Bike {bike_id} is not rented or doesn't exist.")


def display_bikes():
    print("Available Bikes:")
    for bike in bikes:
        if bike.is_available:
            print(f"Bike ID: {bike.bike_id}, Type: {bike.type}, Daily Rate: {bike.get_daily_rate():.2f}/-Rs")


def display_users():
    print("Registered Users:")
    for user in users:
        print(f"User ID: {user.user_id}, Name: {user.password}")

#
while True:
    print("\nBike Rental System Menu:")
    print("1. Add User")
    print("2. Add Bike")
    print("3. Remove Bike")
    print("4. Rent Bike")
    print("5. Return Bike")
    print("6. Display Available Bikes")
    print("7. Display Registered Users")
    print("8. Do visualization")
    print("9. Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        user_id = input("Enter User ID: ")
        password = input("Enter User Password: ")
        age = input("Enter User Age: ")
        add_user(user_id, password, age)


    elif choice == "2":
        bike_id = input("Enter Bike ID: ")
        type = input("Enter Bike Type: ")
        location = input("Enter Station name as A,B,C : ")
        Available = input("Enter available bike: ")
        daily_rate = float(input("Enter Daily Rate: "))
        add_bike(bike_id, type, location, Available, daily_rate)

    elif choice == "3":
        bike_id = input("Enter Bike ID to remove: ")
        remove_bike(bike_id)

    elif choice == "4":
        user_id = input("Enter User ID: ")
        bike_id = input("Enter Bike ID to rent: ")
        season = input("Enter season: ")
        age = input("Enter age: ")
        bikeType = input("Enter bike type: ")
        rental_days = int(input("Enter number of days for rental: "))
        rent_bike(user_id, bike_id, rental_days, age , season, bikeType)

    elif choice == "5":
        user_id = input("Enter User ID: ")
        bike_id = input("Enter Bike ID to return: ")
        return_bike(user_id, bike_id)

    elif choice == "6":
        display_bikes()

    elif choice == "7":
        display_users()

    elif choice == "8":
        c = Analysis()

        print("1. Visualization between bike_type and location: ")
        print("2. Visualization between Bike_type and User_age: ")
        print("3. Visualization between Bike_type and season: ")
        print("4. Visualization Season wise distribution of User age: ")
        print("5. Visualization between Rental_days and Bike Bike_type : ")
        print("6. Correlation matrix of rented bike attribute : ")
        print("7. bike wise availabilty count  : ")

        ch = int(input("Enter your choice: "))

        if ch == 1:
            c.bar_chart1()
        elif ch == 2:
            c.bar_chart2()
        elif ch == 3:
            c.bar_chart3()
        elif ch == 4:
            c.bar_chart4()
        elif ch == 5:
            c.bar_chart5()
        elif ch == 6:
            c.bar_chart6()
        elif ch == 7:
            c.bar_chart7()
        else:
            break

    elif choice == "9":
        print("Exiting the Bike Rental System. Thank you!")
        break

    else:
        print("Invalid choice. Please try again.")
```
